```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import sweetviz as sv
import numpy as np

import sys
sys.path.insert(1, '../utils')
from correlacion.correlacion import remove_highly_correlated_features
```

## Preparar Datos

### Variable respuesta

```python
data_y = pd.read_csv(r"..\data\{}\pmod_base_cruce.csv".format(experiment_name))
data_y.head()
```

```python
print("Cantidad de datos con valores infinitos: ", (data_y.isin([np.inf, -np.inf]).sum()).sum())
print("Cantidad de datos con valores nulos: ", (data_y.isna().sum()).sum())
```

```python
data_y.isna().sum()
```

```python
df_report = data_y[data_y["fecha_var_rpta_alt"] < 202401]
df_report = df_report.drop(columns=["nit_enmascarado", "num_oblig_enmascarado", "num_oblig_orig_enmascarado", "fecha_var_rpta_alt"])
report = sv.analyze([df_report, 'Train'], target_feat='var_rpta_alt')
report.show_html('data_analysis/data_analysis_y.html')
```

```python
data_y["marca_pago"] = data_y["marca_pago"].fillna('NO_DATA')
data_y["marca_pago"] = data_y["marca_pago"].astype('category')

data_y["ajustes_banco"] = data_y["ajustes_banco"].fillna('NO_DATA')
data_y["ajustes_banco"] = data_y["ajustes_banco"].astype('category')

data_y["lote"] = data_y["lote"].fillna(-99)
data_y["lote"] = data_y["lote"].astype('int')
```

```python
data_y.isna().sum()
```

### Saldos diarios

```python
data_sd = pd.read_csv(r"..\data\{}\pmod_tend_sd.csv".format(experiment_name))
data_sd.head()
```

```python
print("Cantidad de valores:", data_sd.shape)
print("Cantidad de datos con valores infinitos: ", (data_sd.isin([np.inf, -np.inf]).sum()).sum())
print("Cantidad de datos con valores nulos: ", (data_sd.isna().sum()).sum())
print("Cantidad de datos con valores negativos: ", ((data_sd < 0).sum()).sum())
```

```python
data_sd = pd.merge(data_sd, 
                   data_y[["nit_enmascarado","num_oblig_enmascarado","num_oblig_orig_enmascarado","fecha_var_rpta_alt",'var_rpta_alt']], 
                   on=["nit_enmascarado","num_oblig_enmascarado","num_oblig_orig_enmascarado","fecha_var_rpta_alt"], 
                   how='inner')
data_sd.head()
```

```python
df_report = data_sd[data_sd["fecha_var_rpta_alt"] < 202401]
df_report = df_report.drop(columns=["nit_enmascarado", "num_oblig_enmascarado", "num_oblig_orig_enmascarado", "fecha_var_rpta_alt"])
report = sv.analyze([df_report, 'Train'], target_feat='var_rpta_alt')
report.show_html('data_analysis/data_analysis_sd.html')
```

```python

data_sd_corr, to_drop = remove_highly_correlated_features(data_sd.drop(columns=["nit_enmascarado", "num_oblig_enmascarado", "num_oblig_orig_enmascarado", "fecha_var_rpta_alt", "var_rpta_alt"]),
                                                          threshold=0.8)
```

```python
data_sd = data_sd.drop(columns=to_drop)
print(data_sd.shape)
data_sd.head()
```

 ### Base modelado

```python
base_0 = pd.merge(data_y, 
         data_sd, 
         on=["nit_enmascarado","num_oblig_enmascarado","num_oblig_orig_enmascarado","fecha_var_rpta_alt", "var_rpta_alt"], 
         how='left')
base_1 = pd.merge(base_0, 
         data_mc, 
         on=["nit_enmascarado","num_oblig_enmascarado","num_oblig_orig_enmascarado","fecha_var_rpta_alt", "var_rpta_alt"], 
         how='left')
base_2 = pd.merge(base_1, 
         data_mcd, 
         on=["nit_enmascarado","num_oblig_enmascarado","num_oblig_orig_enmascarado","fecha_var_rpta_alt", "var_rpta_alt"], 
         how='left')
base_model = pd.merge(base_2, 
            data_pd, 
            on=["nit_enmascarado","num_oblig_enmascarado","num_oblig_orig_enmascarado","fecha_var_rpta_alt", "var_rpta_alt"], 
            how='left')

base_model.head()
```

```python
print("Cantidad de valores:", base_model.shape)
print("Cantidad de datos con valores infinitos: ", (base_model.isin([np.inf, -np.inf]).sum()).sum())
print("Cantidad de datos con valores nulos: ", (base_model.isna().sum()).sum())
```

```python
base_model.isna().sum()
```

```python
df_report = base_model[base_model["fecha_var_rpta_alt"] < 202401]
df_report = df_report.drop(columns=["nit_enmascarado", "num_oblig_enmascarado", "num_oblig_orig_enmascarado", "fecha_var_rpta_alt"])
report = sv.analyze([df_report, 'Train'], target_feat='var_rpta_alt')
report.show_html('data_analysis/base_model.html')
```

```python
base_model.to_csv(r"..\data\{}\pmod_base_model.csv".format(experiment_name), index=False)
```
